```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
%matplotlib inline
```

```python
colour_vecs = np.concatenate([grid.reshape((-1,1)) for grid in np.meshgrid(
        *(np.arange(0,256,16) for _ in range(3))
        )],axis=1)
colours = ["#%02x%02x%02x" % tuple(vec) for vec in colour_vecs]
```

```python
len(colour_vecs)
```

```python
tsne = TSNE(n_components=2)
colours_2d = tsne.fit_transform(colour_vecs)
```

```python
plt.scatter(colours_2d[:,0],colours_2d[:,1],c=colours,s=5)
```
